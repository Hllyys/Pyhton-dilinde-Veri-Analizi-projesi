```python
# Eksik veri analizi için bir işlev
def eksik_veri_analizi(df):
    # Her sütundaki eksik veri sayısını hesapla
    eksik_veri_sayisi = df.isnull().sum()
    # Her sütundaki eksik veri oranını hesapla
    eksik_veri_orani = (df.isnull().sum() / len(df)) * 100
    # Eksik veri analizini bir veri çerçevesinde topla
    eksik_veri_analizi = pd.DataFrame({'Eksik Veri Sayısı': eksik_veri_sayisi, 'Eksik Veri Oranı (%)': eksik_veri_orani})
    return eksik_veri_analizi
```

```python
# Ayrık veri analizi için bir işlev
def ayrık_veri_analizi(df):
    # Veri setindeki sayısal sütunlar için temel istatistiksel özet bilgileri al
    ayrık_veri_analizi_df = df.describe().transpose()
    return ayrık_veri_analizi_df
```

```python
# KNN modelleme için bir işlev
def knn_modelleme(X_train, X_test, y_train, y_test):
    # KNN için hiperparametre optimizasyonu yapılacak parametrelerin bir listesini oluştur
    param_grid = {'n_neighbors': [3, 5, 7],
                  'weights': ['uniform', 'distance'],
                  'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}
    # KNN modelini oluştur
    knn_model = KNeighborsClassifier()
    # Parametre optimizasyonunu yap
    grid_search = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    # En iyi parametreleri al
    best_knn_params = grid_search.best_params_
    # En iyi modeli oluştur
    best_knn_model = KNeighborsClassifier(**best_knn_params)
    # En iyi modeli eğit
    best_knn_model.fit(X_train, y_train)
    # Test seti üzerinde tahmin yap
    y_pred_knn = best_knn_model.predict(X_test)
    # Doğruluk (accuracy) hesapla
    accuracy_knn = accuracy_score(y_test, y_pred_knn)
    # Karışıklık matrisi (confusion matrix) hesapla
    confusion_matrix_knn = confusion_matrix(y_test, y_pred_knn)
    # Sınıflandırma raporu (classification report) hesapla
    classification_report_knn = classification_report(y_test, y_pred_knn, zero_division=1)
    return best_knn_model, accuracy_knn, confusion_matrix_knn, classification_report_knn

```

```python
# Karar ağacı modelleme için bir işlev
def decision_tree_modelleme(X_train, X_test, y_train, y_test):
    # Karar ağacı için hiperparametre optimizasyonu yapılacak parametrelerin bir listesini oluştur
    param_grid = {'criterion': ['gini', 'entropy'],
                  'splitter': ['best', 'random'],
                  'max_depth': [None, 5, 10, 15, 20],
                  'min_samples_split': [2, 5, 10],
                  'min_samples_leaf': [1, 2, 4]}
    # Karar ağacı modelini oluştur
    dt_model = DecisionTreeClassifier()
    # Parametre optimizasyonunu yap
    grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    # En iyi parametreleri al
    best_dt_params = grid_search.best_params_
    # En iyi modeli oluştur
    best_dt_model = DecisionTreeClassifier(**best_dt_params)
    # En iyi modeli eğit
    best_dt_model.fit(X_train, y_train)
    # Test seti üzerinde tahmin yap
    y_pred_dt = best_dt_model.predict(X_test)
    # Doğruluk (accuracy) hesapla
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    # Karışıklık matrisi (confusion matrix) hesapla
    confusion_matrix_dt = confusion_matrix(y_test, y_pred_dt)
    # Sınıflandırma raporu (classification report) hesapla
    classification_report_dt = classification_report(y_test, y_pred_dt, zero_division=1)  
    return best_dt_model, accuracy_dt, confusion_matrix_dt, classification_report_dt
```

```python
# Rastgele orman modelleme için bir işlev
def random_forest_modelleme(X_train, X_test, y_train, y_test):
    # Rastgele orman için hiperparametre optimizasyonu yapılacak parametrelerin bir listesini oluştur
    param_grid = {'n_estimators': [100, 200, 300],
                  'criterion': ['gini', 'entropy'],
                  'max_depth': [None, 10, 20, 30, 50]}
    # Rastgele orman modelini oluştur
    rf_model = RandomForestClassifier()
    # Parametre optimizasyonunu yap
    grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    # En iyi modeli al
    best_rf_model = grid_search.best_estimator_
    # En iyi modeli eğit
    best_rf_model.fit(X_train, y_train)
    # Test seti üzerinde tahmin yap
    y_pred_rf = best_rf_model.predict(X_test)
    # Doğruluk (accuracy) hesapla
    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    # Karışıklık matrisi (confusion matrix) hesapla
    confusion_matrix_rf = confusion_matrix(y_test, y_pred_rf)
    # Sınıflandırma raporu (classification report) hesapla
    classification_report_rf = classification_report(y_test, y_pred_rf, zero_division=1)
    return best_rf_model, accuracy_rf, confusion_matrix_rf, classification_report_rf
```

```python
# Kümeleme için Siluet skoru hesaplama işlevi
def silhouette_scoru_kumeleme(X):
    # KMeans modelini oluştur
    kmeans_model = KMeans(n_clusters=2)
    # Veriyi kümele
    kmeans_model.fit(X)
    # Etiketleri al
    labels = kmeans_model.labels_
    # Siluet skorunu hesapla
    silhouette_score_kmeans = silhouette_score(X, labels)
    return silhouette_score_kmeans

```

```python
def veri_isleme_ve_modelleme(df):
# Eksik veri analizi yapar ve eksik verileri işler.
# Aykırı değerlerin tespit edilmesi ve düzeltilmesi için box plotlar kullanır.
# Veri seti modelleme işlemleri için hazırlanır:
# Kategorik sütunlar sayısal değerlere dönüştürülür.
# Bağımsız ve bağımlı değişkenler ayrıştırılır.
# Veri normalizasyonu yapılır.
# Eğitim ve test setleri oluşturulur.
# KNN (K En Yakın Komşu), Karar Ağacı ve Rastgele Orman modelleri uygulanır ve bu modellerin performansları değerlendirilir.
# Veri boyutunu azaltmak için PCA (Temel Bileşen Analizi) ve kümeleme (KMeans) yöntemleri kullanılır.
# Sınıflandırma algoritmalarının performanslarını karşılaştırmak için Friedman Testi uygulanır.

    # Eksik veri analizi ve işlemleri
    eksik_veri_analizi_df = eksik_veri_analizi(df)
    print("Eksik Veri Analizi:")
    print(eksik_veri_analizi_df)
    
    ayrık_veri_analizi_df = ayrık_veri_analizi(df)
    print("Ayrık Veri Analizi:")
    print(ayrık_veri_analizi_df)

    plt.figure(figsize=(10, 6))
    plt.bar(eksik_veri_analizi_df.index, eksik_veri_analizi_df['Eksik Veri Sayısı'])
    plt.title("Sütunlara Göre Eksik Veri Sayısı")
    plt.xlabel("Sütunlar")
    plt.ylabel("Eksik Veri Sayısı")
    plt.xticks(rotation=45)
    plt.show()

    # Eksik verilerin önişlenmesi
    df.dropna(inplace=True)
    # Eksik verileri ortalama ile doldurma
    df['Sipariş Durum'] = df['Sipariş Durum'].fillna(df['Sipariş Durum'].mode()[0])
    df['Ödeme Tutarı'] = df['Ödeme Tutarı'].fillna(df['Ödeme Tutarı'].mean())
    # Box plot ile aykırı değerleri gösterme (ayrı ayrı çizdirme)
    plt.figure(figsize=(10, 6))
    
    # Ürün Fiyatı için box plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df['Ürün Fiyatı'])
    plt.title("Ürün Fiyatı Aykırı Değerlerin Box Plot Gösterimi")

    # Sipariş Toplam Tutarı için box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df['Sipariş Toplam Tutar'])
    plt.title("Sipariş Toplam Tutarı Aykırı Değerlerin Box Plot Gösterimi")

    plt.tight_layout()  # Grafiklerin örtüşmemesi için
    plt.show()     
    # Aykırı değerleri düzeltme
    df_filtered = df.copy()
    Q1 = df_filtered['Ürün Fiyatı'].quantile(0.25)
    Q3 = df_filtered['Ürün Fiyatı'].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    df_filtered = df_filtered[(df_filtered['Ürün Fiyatı'] > alt_sinir) & (df_filtered['Ürün Fiyatı'] < ust_sinir)]

    Q1 = df_filtered['Sipariş Toplam Tutar'].quantile(0.25)
    Q3 = df_filtered['Sipariş Toplam Tutar'].quantile(0.75)
    IQR = Q3 - Q1
    alt_sinir = Q1 - 1.5 * IQR
    ust_sinir = Q3 + 1.5 * IQR
    df_filtered = df_filtered[(df_filtered['Sipariş Toplam Tutar'] > alt_sinir) & (df_filtered['Sipariş Toplam Tutar'] < ust_sinir)]

    # Box plot ile düzeltilmiş aykırı değerleri gösterme
    plt.figure(figsize=(10, 6))

    # Ürün Fiyatı için box plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x=df_filtered['Ürün Fiyatı'])
    plt.title("Ürün Fiyatı Aykırı Değerlerin Düzeltildikten Sonra Box Plot Gösterimi")

    # Sipariş Toplam Tutarı için box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df_filtered['Sipariş Toplam Tutar'])
    plt.title("Sipariş Toplam Tutarı Aykırı Değerlerin Düzeltildikten Sonra Box Plot Gösterimi")

    plt.tight_layout()  # Grafiklerin örtüşmemesi için
    plt.show()

    # Aykırı değerleri düzeltilmiş veri setini tekrar kullanmak üzere döndürme
    df = df_filtered

    # Modelleme işlemleri
    # LabelEncoder nesnesini oluşturalım
    label_encoder = LabelEncoder()

    # 'Sipariş Durum' sütununu sayısal değerlere dönüştürelim
    df['Sipariş Durum Num'] = label_encoder.fit_transform(df['Sipariş Durum'])

    # Bağımsız değişkenler (X) ve bağımlı değişken (y) ayrıştırma
    X = df.drop(columns=['Sipariş Durum', 'Sipariş Durum Num'])
    y = df['Sipariş Durum Num']

    # Normalizasyon işlemi
    scaler = MinMaxScaler()
    X_normalized = scaler.fit_transform(X)

    # Eğitim ve test setlerine ayırma
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

    # KNN, Karar Ağacı ve Rastgele Orman modellerini eğitme ve performanslarını değerlendirme
    print("KNN Modelleme Sonuçları:")
    knn_model, knn_accuracy, knn_report, knn_conf_matrix = knn_modelleme(X_train, X_test, y_train, y_test)
    print("Accuracy:", knn_accuracy)
    print("Classification Report:")
    print(knn_report)
    print("Confusion Matrix:")
    print(knn_conf_matrix)

    print("\nDecison Tree Modelleme Sonuçları:")
    decision_tree_model, decision_tree_accuracy, decision_tree_report, decision_tree_conf_matrix = decision_tree_modelleme(X_train, X_test, y_train, y_test)
    print("Accuracy:", decision_tree_accuracy)
    print("Classification Report:")
    print(decision_tree_report)
    print("Confusion Matrix:")
    print(decision_tree_conf_matrix)

    print("\nRandom Forest Modelleme Sonuçları:")
    random_forest_model, random_forest_accuracy, random_forest_report, random_forest_conf_matrix = random_forest_modelleme(X_train, X_test, y_train, y_test)
    print("Accuracy:", random_forest_accuracy)
    print("Classification Report:")
    print(random_forest_report)
    print("Confusion Matrix:")
    print(random_forest_conf_matrix)

     # Doğruluk (accuracy) skorlarını görselleştirme
    models = ['KNN', 'Decision Tree', 'Random Forest']
    accuracy_scores = [knn_accuracy, decision_tree_accuracy, random_forest_accuracy]

    plt.figure(figsize=(10, 6))
    plt.bar(models, accuracy_scores, color=['skyblue', 'lightgreen', 'lightcoral'])
    plt.title('Model Doğruluk Skorları')
    plt.xlabel('Modeller')
    plt.ylabel('Doğruluk Skoru')
    plt.ylim(0.0, 1.0)  # Doğruluk skorları aralığı
    plt.show()

    # PCA ve Kümeleme (Clustering) işlemleri
    X = df[['Ürün Fiyatı', 'Sipariş Toplam Tutar']]

    # Boyut azaltma işlemi için PCA uygulayalım
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # KMeans algoritması ile kümeleme yapalım
    kmeans_model = KMeans(n_clusters=2, random_state=42, n_init=10)
    kmeans_model.fit(X_pca)
    y_kmeans = kmeans_model.predict(X_pca)

    # Küme merkezlerini alalım
    centroids = kmeans_model.cluster_centers_

    # Kümeleme sonuçlarını görselleştirelim
    plt.figure(figsize=(10, 6))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, s=50, cmap='viridis')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5)
    plt.title('K-Means Kümeleme Sonuçları (PCA ile Boyut Azaltılmış)')
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    plt.show()

    # Siluet skoru hesaplayalım
    siluet_scoru = silhouette_score(X_pca, y_kmeans)
    print("Kümeleme (Clustering) İşlemi Siluet Skoru:", siluet_scoru)

    # Friedmann Testi
    # Sınıflandırma algoritmalarının performanslarını karşılaştıralım
    knn_acc_scores = []
    dt_acc_scores = []
    rf_acc_scores = []

    
    X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2)

    _, accuracy_knn, _, _ = knn_modelleme(X_train, X_test, y_train, y_test)
    _, accuracy_dt, _, _ = decision_tree_modelleme(X_train, X_test, y_train, y_test)
    _, accuracy_rf, _, _ = random_forest_modelleme(X_train, X_test, y_train, y_test)

    knn_acc_scores.append(accuracy_knn)
    dt_acc_scores.append(accuracy_dt)
    rf_acc_scores.append(accuracy_rf)

    friedman_test_statistic, p_value = friedmanchisquare(knn_acc_scores, dt_acc_scores, rf_acc_scores)
    print("\nSınıflandırma Algoritmalarının Performans Karşılaştırması:")
    print("KNN Ortalama Doğruluk Skoru:", np.mean(knn_acc_scores))
    print("Decision Tree Ortalama Doğruluk Skoru:", np.mean(dt_acc_scores))
    print("Random Forest Ortalama Doğruluk Skoru:", np.mean(rf_acc_scores))

    print("\nFriedman Testi İstatistiği:", friedman_test_statistic)
    print("p-değeri:", p_value)

    if p_value < 0.05:
        print("p-değeri anlamlı düzeyde küçük olduğu için gruplar arasında istatistiksel olarak anlamlı bir fark vardır.")
    else:
        print("p-değeri 0.05'ten büyük olduğu için gruplar arasında istatistiksel olarak anlamlı bir fark yoktur.")

veri_isleme_ve_modelleme(df)
```
